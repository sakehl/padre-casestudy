```python
def data_points(data: str)->List[float]:
    xs: List[float] = list(float(l.split()[1]) for l in data.split('\n') if l != "" and not l.startswith("Silicon"))
    return xs

def out_of_range(data: str, maxi: float, op: str, offset: int=0)->List[Tuple[float, int, int]]:
    xs: List[float] = data_points(data)
    res: List[Tuple[float, int, int]] = list(((d, i, int(maxi)) for i,d in enumerate(xs) if d>maxi))
        
    for i,x,y in res:
        plt.plot(x,180-1, op) 
        plt.annotate('%d' % (i), xy=(x,y-10-offset))
        
    return res
```
